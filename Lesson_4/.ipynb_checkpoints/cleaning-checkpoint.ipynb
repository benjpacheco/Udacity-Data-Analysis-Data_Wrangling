### Quality

#### patients table

- zip code is a float not a string
- zip code has four digits soemtimes
- Tim Neudorf height is 27 in instead of 72 in
- full state names sometimes, abbreviations other times
- Dsvid Gustafsson
- Missing demographic information (address - contact columns)
- Erroneous datatypes (assigned sex, state, zip_code, and birthdate columns)
- Multiple phone number formats
- Default John Doe data
- Multiple records for Jakobsen, Gersten, Taylor
- kgs instead of lbs for Zaitseva weight

#### treatments table

- missing HbA1c changes
- the letter u in starting and ending doses for Auralin and Novodra
- lowercase given names and surnames
- missing records (280 instead of 350)
- Erroneous datatypes (auralin and novodra columns)
- Inaccurate HbA1c changes (4s mistaken as 9s)
- Nulls represented as dashes (-) in auralin and novodra columns)

#### adverse_reactions table

- lowercase given names and surnames

### Tidiness

#### patients table

- contact column should be split into phone number and email

#### treatments table

- three variables in two columns (treatment, start dose, end dose)

#### adverse_reactions table

- adverse_reaction column should be joined with treatments table, it is the same type of observational unit
- other columns should be removed since they are already in patient and treatment tables
